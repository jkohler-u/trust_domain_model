```python
import plotting
import pandas as pd
```

```python
#open csv
gt_df = pd.read_csv("gt_df.csv")

# Create a dictionary to hold the arrays for each speaker
speaker_data = {}

# Loop through every unique speaker in the dataframe
for speaker in gt_df['speaker'].unique():
    # 1. Filter the dataframe for the current speaker
    speaker_subset = gt_df[gt_df['speaker'] == speaker]
    
    # 2. Extract the columns as numpy arrays
    truth_array = speaker_subset['is_true'].values
    domain_array = speaker_subset['domain'].values
    
    # 3. Store them in the dictionary as a tuple or another dictionary
    speaker_data[speaker] = {
        'is_true': truth_array,
        'domain': domain_array
    }

```
